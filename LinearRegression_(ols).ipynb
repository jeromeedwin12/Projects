```python
from sklearn.decomposition import PCA
pca = PCA(n_components=150)
principalComponents = pca.fit_transform(data1)
```

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(principalComponents)

y=df["Property_Sale_Price"]
model = sm.OLS(y, X).fit()
summary = model.summary()
print(summary)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(principalComponents,y,test_size=0.2,random_state=123)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
